# Sample Mean and the Central Limit Theorem (CLT)

The **sample mean** is a fundamental statistic in data analysis, and its behavior is described by the **Central Limit Theorem (CLT)**â€”a cornerstone of inferential statistics.

## What is the Central Limit Theorem?

The **Central Limit Theorem (CLT)** states:

> If you take sufficiently large random samples from any population (regardless of the population's distribution), the distribution of the sample means will approach a normal (bell-shaped) distribution as the sample size increases.

## Key Points

- The **mean** of the sampling distribution of the sample mean ($\bar{x}$) equals the **population mean** ($\mu$).
- The **standard deviation** of the sampling distribution (the **standard error**) is:
  $$
  \text{Standard Error} = \frac{\sigma}{\sqrt{n}}
  $$
  where $\sigma$ is the population standard deviation and $n$ is the sample size.
- As **sample size increases**, the distribution of sample means becomes more tightly clustered around the population mean and more closely resembles a normal distribution.

## Why is the CLT Important?

- **Normality:** It allows us to use normal probability calculations for sample means, even if the underlying data is not normal, provided the sample size is large enough (commonly $n \geq 30$).
- **Inference:** It justifies using the sample mean as an estimator for the population mean and supports confidence intervals and hypothesis testing.

## Python Example: Simulating the Central Limit Theorem

Below is an example showing how the distribution of sample means becomes normal as sample size increases, even when sampling from a non-normal population (exponential distribution):


```python

import numpy as np
import matplotlib.pyplot as plt

# Population: Exponential distribution (not normal)
population = np.random.exponential(scale=2, size=100000)

sample_sizes = [5, 30, 100]
plt.figure(figsize=(15, 4))

for i, n in enumerate(sample_sizes, 1):
    sample_means = []
    # Draw 10,000 samples, each of size n, and compute their means
    for _ in range(10000):
        sample = np.random.choice(population, size=n, replace=False)
        sample_means.append(np.mean(sample))
    plt.subplot(1, 3, i)
    plt.hist(sample_means, bins=30, density=True, alpha=0.7, color='skyblue', edgecolor='black')
    plt.title(f'Sample Size = {n}')
    plt.xlabel('Sample Mean')
    plt.ylabel('Density')

plt.suptitle('Central Limit Theorem: Distribution of Sample Means')
plt.tight_layout()
plt.show()
```



## What Does This Show?

- **Small sample size (n=5):** The distribution of sample means is still skewed, reflecting the original population.
- **Larger sample sizes (n=30, n=100):** The distribution of sample means becomes more normal and centered around the population mean.

## Summary

- The **sample mean** is a key statistic for summarizing data.
- The **Central Limit Theorem** explains why the distribution of sample means is approximately normal for large samples, regardless of the population's distribution.
- This property makes the sample mean a powerful tool for statistical inference.
